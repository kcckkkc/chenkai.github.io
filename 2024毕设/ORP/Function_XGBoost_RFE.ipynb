```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import KFold
import optuna

# 
train_data = pd.read_excel(r'C:\Users\k\Desktop\\ORP\ORP4\XGBoost\ORP_function_train.xlsx')
test_data = pd.read_excel(r'C:\Users\k\Desktop\\ORP\ORP4\XGBoost\ORP_function_test.xlsx')

# 
X_train, y_train = train_data.drop(['E','File'], axis=1), train_data['E']
X_test, y_test = test_data.drop(['E','File'], axis=1), test_data['E']

def objective(trial):
    param = {
        'tree_method': 'hist',  # 
        'device': 'cpu',       # CUDAGPU
        'learning_rate': trial.suggest_float('learning_rate', 0.001, 1),
        'n_estimators': trial.suggest_int('n_estimators', 100, 500),
        'max_depth': trial.suggest_int('max_depth', 2, 5),
        'min_child_weight': trial.suggest_int('min_child_weight', 1, 5),
        'gamma': trial.suggest_float('gamma', 0, 0.6),
        'subsample': trial.suggest_float('subsample', 0.6, 1.0),
        'colsample_bytree': trial.suggest_float('colsample_bytree', 0.6, 1.0)
    }

    model = XGBRegressor(**param)
    
    # k
    kf = KFold(n_splits=5, shuffle=True, random_state=42)
    rmse_scores = []

    for train_index, val_index in kf.split(X_train):
        X_train_fold, X_val_fold = X_train.iloc[train_index], X_train.iloc[val_index]
        y_train_fold, y_val_fold = y_train.iloc[train_index], y_train.iloc[val_index]

        model.fit(X_train_fold, y_train_fold)
        y_val_pred = model.predict(X_val_fold)
        rmse_fold = np.sqrt(mean_squared_error(y_val_fold, y_val_pred))
        rmse_scores.append(rmse_fold)
    
    # RMSE
    avg_rmse = np.mean(rmse_scores)
    
    return avg_rmse

# Optuna
study = optuna.create_study(direction='minimize')
study.optimize(objective, n_trials=1000)

# 
best_params = study.best_params

def calculate_adjusted_r2(r_squared, n, p):
    return 1 - (1 - r_squared) * (n - 1) / (n - p - 1)

# 
average_rmses = []
average_adjusted_r2s = []
test_rmses = []
test_adjusted_r2s = []

features = X_train.columns.tolist()
while len(features) > 2:
    # 
    model = XGBRegressor(**best_params)
    model.fit(X_train[features], y_train)
    feature_importances = model.feature_importances_

    # 
    sorted_idx = np.argsort(feature_importances)
    features.pop(sorted_idx[0])
# KRMSER2
    kf = KFold(n_splits=5, shuffle=True, random_state=42)
    rmse_scores = []
    r2_scores = []
    for train_index, val_index in kf.split(X_train[features]):
        X_train_fold, X_val_fold = X_train[features].iloc[train_index], X_train[features].iloc[val_index]
        y_train_fold, y_val_fold = y_train.iloc[train_index], y_train.iloc[val_index]

        model.fit(X_train_fold, y_train_fold)
        y_val_pred = model.predict(X_val_fold)
        rmse_scores.append(np.sqrt(mean_squared_error(y_val_fold, y_val_pred)))
        r2_scores.append(r2_score(y_val_fold, y_val_pred))

    avg_rmse = np.mean(rmse_scores)
    avg_r2 = np.mean(r2_scores)
    avg_adjusted_r2 = calculate_adjusted_r2(avg_r2, len(X_train_fold), len(features))

    # RMSER2
    y_test_pred = model.predict(X_test[features])
    test_rmse = np.sqrt(mean_squared_error(y_test, y_test_pred))
    test_r2 = r2_score(y_test, y_test_pred)
    test_adjusted_r2 = calculate_adjusted_r2(test_r2, len(y_test), len(features))

    average_rmses.append(avg_rmse)
    average_adjusted_r2s.append(avg_adjusted_r2)
    test_rmses.append(test_rmse)
    test_adjusted_r2s.append(test_adjusted_r2)


# 
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(range(len(average_rmses)), average_rmses, marker='o')
plt.xlabel('Number of Features Removed')
plt.ylabel('Average RMSE')
plt.title('RMSE vs. Number of Features')

plt.subplot(1, 2, 2)
plt.plot(range(len(average_adjusted_r2s)), average_adjusted_r2s, marker='o', color='green')
plt.xlabel('Number of Features Removed')
plt.ylabel('Average Adjusted R2')
plt.title('Adjusted R2 vs. Number of Features')

plt.tight_layout()
plt.show()
```
