```python
import pandas as pd
import matplotlib.pyplot as plt
from math import pi

# 
data = {
    "Model": ["ENR", "ANN", "RFR", "SVR", "XGBoost"],
   "$R^{2}_{\t{train}}$": [0.73,0.98,0.96,0.94,0.96],
    "$1/RMSE_{\t{train}}$": [0,1,0.52,0.35,0.52],
    "$1/MAPE_{\t{train}}$": [0,1,0.46,0.43,0.49],
    "$R^{2}_{\t{test}}$": [0.65,0.77,0.69,0.79,0.65],
    "$1/RMSE_{\t{test}}$": [0,0.78,0.23,1,0.07],
    "$1/MAPE_{\t{test}}$": [0,0.71,0.26,1,0.14]
}

#  DataFrame
df = pd.DataFrame(data)

# 
categories = list(df.columns[1:])
N = len(categories)

# 
angles = [n / float(N) * 2 * pi for n in range(N)]
angles += angles[:1]

# 
plt.figure(figsize=(8, 8))
ax = plt.subplot(111, polar=True)

# 
plt.xticks(angles[:-1], categories,fontproperties = 'Arial', size = 16)

#  y 
ax.set_rlabel_position(0)
plt.yticks([0.25, 0.5, 0.75], ["0.25","0.50","0.75"], color="grey", size=8,fontproperties = 'Arial')
plt.ylim(0,1)

# 
for i in range(len(df)):
    values = df.loc[i].drop('Model').values.flatten().tolist()
    values += values[:1]
    ax.plot(angles, values, linewidth=1, linestyle='solid', label=df.loc[i, 'Model'])
    ax.fill(angles, values, alpha=0.1)

# 

plt.legend(prop={'family' : 'Arial', 'size'   : 12},bbox_to_anchor=(1,1))
plt.rcParams['figure.dpi']=600


# 
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import shap

# 
np.random.seed(0)
data = np.random.rand(100, 5)
model_output = data @ np.array([1.5, -2., 0.5, 0., 3.]) + np.random.normal(0, 0.1, 100)

#  SHAP  SHAP 
explainer = shap.Explainer(lambda x: x @ np.array([1.5, -2., 0.5, 0., 3.]), data)
shap_values = explainer(data)

#  Summary Plot
plt.figure(figsize=(10, 6))
shap.summary_plot(shap_values, data,show=False)
plt.title("SHAP Summary Plot")
plt.tight_layout()
plt.show()

#  Beeswarm Plot
plt.figure(figsize=(10, 6))
shap.summary_plot(shap_values, data, show=False)
plt.title("SHAP Beeswarm Plot")
plt.tight_layout()
plt.show()
```
