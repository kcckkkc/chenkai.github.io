```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import statsmodels.api as sm

# 假设您的数据在Excel文件中，您可以使用pandas来读取
# 替换'your_file.xlsx'和'sheet_name'为您的文件名和工作表名
df = pd.read_excel('C:/Users/k/Desktop/硕士论文/SO4/SO4_240301/k_ORP/k_ORP.xlsx', sheet_name='Sheet1')

# 假设您的两列数据分别命名为'X'和'Y'
X = df['E']
Y = df['lgk']

# 创建散点图
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['figure.dpi']=600
bwith = 5 #边框宽度设置为2
plt.figure(figsize=(8, 8))
sns.scatterplot(x=X, y=Y,color='black',edgecolor='none')

# 计算拟合曲线（线性回归）
X_sm = sm.add_constant(X)  # 添加常数项
model = sm.OLS(Y, X_sm)    # 创建模型
results = model.fit()      # 拟合模型

# 获取拟合线的值
X_fit = np.linspace(0.5, 5.5, 100)
Y_fit = results.params[0] + results.params[1] * X_fit

# 添加拟合线到散点图上
plt.plot(X_fit, Y_fit, color='red')
# 计算R²值
R2 = results.rsquared

# 添加拟合表达式和R²值到图形上
plt.text(4, 7, f'y = {results.params[1]:.2f}x + 11.88\n$R^2$ = {R2:.2f}', color='red', fontproperties = 'Times New Roman',fontsize=16,linespacing=1.5,weight='bold')

plt.text(3.2, 8, f'(3.10,7.66)', color='green', fontproperties = 'Times New Roman',fontsize=16,linespacing=1.5,weight='bold')

plt.text(2.6, 8.6, f'(2.50,8.48)', color='blue', fontproperties = 'Times New Roman',fontsize=16,linespacing=1.5,weight='bold')
# 添加标题和标签
#plt.title('散点图与拟合曲线')
plt.xlim(0.5,5.5)
plt.ylim(3,12)
#plt.xticks(np.arange(0.5, 5.5, 0.5))
plt.yticks(fontproperties = 'Arial', size = 16)
# 设置 x 轴刻度间隔为 0.5
plt.xticks(np.arange(0.5, 6, 0.5))

plt.xticks(fontproperties = 'Arial', size = 16)

plt.xlabel(r'$E_{ox}^{1}$(V vs.SHE)',fontdict={'family': 'Arial', 'size': 18})
plt.ylabel('lgk',fontdict={'family': 'Arial', 'size': 18})
# 在x=2.5处添加水平线和垂直线
# 计算两个特殊点的y值
y_at_2_5 = results.params[0] + results.params[1] * 2.5
y_at_3_1 = results.params[0] + results.params[1] * 3.1

# 在x=2.5处添加连接x轴和y轴的线
plt.plot([2.5, 2.5], [0, y_at_2_5], color='blue', linestyle='--') # 连接x轴的线
plt.plot([0, 2.5], [y_at_2_5, y_at_2_5], color='blue', linestyle='--') # 连接y轴的线

# 在x=3.1处添加连接x轴和y轴的线
plt.plot([3.1, 3.1], [0, y_at_3_1], color='green', linestyle='--') # 连接x轴的线
plt.plot([0, 3.1], [y_at_3_1, y_at_3_1], color='green', linestyle='--') # 连接y轴的线


# 显示图形
plt.show()
```
