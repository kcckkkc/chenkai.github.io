```python
import os
import shutil

# 源文件夹路径
source_folder = 'D:/pubchem/聚类分析/sdf'

# 目标文件夹路径
target_folder = 'D:/pubchem/聚类分析/sdf_270'

# 读取包含文件名的txt文档
with open('D:/pubchem/聚类分析/CID.txt', 'r') as file:
    file_names = [line.strip() for line in file.readlines()]


# 遍历文件名列表
for file_name in file_names:
    # 构建源文件路径
    source_file = os.path.join(source_folder, file_name)
    
    # 构建目标文件路径
    target_file = os.path.join(target_folder, file_name)
    
    # 检查文件是否存在并复制文件
    if os.path.exists(source_file):
        shutil.copyfile(source_file, target_file)
        print(f'文件 {file_name} 已成功复制到目标文件夹')
    else:
        print(f'文件 {file_name} 不存在于源文件夹')

print('批量文件提取完成')
```
